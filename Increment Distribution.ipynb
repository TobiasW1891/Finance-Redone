```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

sns.set()
```

```python
kernel = 5 # for loading data
```

```python
RealData = np.load("MarketCorrWeekly.npy")
plt.plot(RealData)
```

```python
SimulatedMean = np.load("simulatedDataX_B10_K"+str(kernel)+"_mean.npy")
SimulatedMAP = np.load("simulatedDataX_B10_K"+str(kernel)+"_MAP.npy")


sns.kdeplot(RealData, label="Real")
plt.xlabel("Market Correlation")
#sns.kdeplot(SimulatedMean, label="Mean")
#sns.kdeplot(SimulatedMAP, label="MAP")
plt.title("Time Series KDE")
plt.legend()
plt.show()
```

```python
plt.hist(SimulatedMean, density= True, label="Simulation Mean")
plt.hist(SimulatedMAP, density= True, label="Simulation MAP", color="green", alpha = 0.7)
plt.hist(RealData, density= True, color="red", alpha = 0.5,label="Real")
plt.legend()
plt.title("Historgram of Time Series")
plt.show()
```

## First Increment

```python
IncrementMean = SimulatedMean[1:] - SimulatedMean[:-1]
IncrementMAP = SimulatedMAP[1:] - SimulatedMAP[:-1]
IncrementReal = RealData[1:] - RealData[:-1]


#plt.hist(IncrementMean, density= True, label="Simulation Mean")
plt.hist(IncrementMAP, density= True, label="Simulation MAP", color="green", alpha = 0.7)
plt.hist(IncrementReal, density= True, color="red", alpha = 0.5, label="Real")
plt.legend()
plt.title("Historgram of Increments")
plt.show()
```

```python

sns.kdeplot(IncrementReal, label="Real")
plt.xlabel("First Increment of Market Correlation")
#sns.kdeplot(IncrementMean, label="Mean")
#sns.kdeplot(IncrementMAP, label="MAP")
plt.title("Increment Distribution KDE")
plt.legend()
plt.show()
```

```python
Increment2Mean = SimulatedMean[2:] - SimulatedMean[:-2]
Increment2MAP = SimulatedMAP[2:] - SimulatedMAP[:-2]
Increment2Real = RealData[2:] - RealData[:-2]


plt.hist(Increment2Mean, density= True, label="Simulation Mean")
plt.hist(Increment2MAP, density= True, label="Simulation MAP", color="green", alpha = 0.7)
plt.hist(Increment2Real, density= True, color="red", alpha = 0.5, label="Real")
plt.legend()
plt.title("Historgram of 2nd Increments")
plt.show()
```

```python

sns.kdeplot(Increment2Real, label="Real")
plt.xlabel("Second Order Increment of Market Correlation")
#sns.kdeplot(Increment2Mean, label="Mean")
#sns.kdeplot(Increment2MAP, label="MAP")
plt.title("Second Order Increment Distribution KDE")
plt.legend()
plt.show()
```

## Now all in one plot

```python
fig, axes = plt.subplots(figsize=(11,4), ncols=3, nrows=1)
print(axes, axes.shape)
```

```python
fig, axes = plt.subplots(figsize=(9,3.5), ncols=3, nrows=1)

sns.kdeplot(RealData, label="Real", ax = axes[0])
axes[0].set_xlim([0, 1])
axes[0].set_xlabel("Market Correlation")
sns.kdeplot(SimulatedMean, label="Mean", ax = axes[0])
sns.kdeplot(SimulatedMAP, label="MAP", ax = axes[0])
#axes[0].set_title("Time Series KDE")
axes[0].legend()



sns.kdeplot(IncrementReal, label="Real", ax = axes[1])
axes[1].set_xlabel("First Order Increments")
sns.kdeplot(IncrementMean, label="Mean", ax = axes[1])
sns.kdeplot(IncrementMAP, label="MAP", ax = axes[1])
#axes[1].set_title("Increment Distribution KDE")
axes[1].legend()



sns.kdeplot(IncrementReal, label="Real", ax = axes[2])
axes[2].set_xlabel("Second Order Increments")
sns.kdeplot(IncrementMean, label="Mean", ax = axes[2])
sns.kdeplot(IncrementMAP, label="MAP", ax = axes[2])
#axes[2].set_title("Second Increment Distribution KDE")
axes[2].legend()

plt.tight_layout()
fig.suptitle("KDE of Market Correlation and First Two Increments with Kernel "+str(kernel), fontsize = 14)
fig.savefig("Distributions_Real_vs_Simulation_K"+str(kernel)+".pdf")
fig.show()
```
